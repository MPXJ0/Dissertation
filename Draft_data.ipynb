```python
def post_elexon(url):
    http_obj = httplib2.Http()
    resp, content = http_obj.request(
        uri=url,
        method='GET',
        headers={'Content-Type': 'application/xml; charset=UTF-8'},
    )
    print('===Response===')
    print(pformat(resp))
    print('===Content===')
    print(pformat(content))
    print('===Finished===')
def main():
    post_elexon(
        url='https://api.bmreports.com/BMRS/B1770/v1?APIKey=kk73d4dvvfdgjd5&SettlementDate=2019-01-01&Period=13&ServiceType=xml',
    )
    
if __name__ == "__main__":
    main()
```
